```python
# From https://sirinnes.wordpress.com/2013/04/25/pareto-frontier-graphic-via-python/
def plot_pareto_frontier(Xs, Ys, maxX=True, maxY=True, xlabel='X label', ylabel='Y label'):
    '''Pareto frontier selection process'''
    sorted_list = sorted([[Xs[i], Ys[i]] for i in range(len(Xs))], reverse=maxY)
    pareto_front = [sorted_list[0]]
    for pair in sorted_list[1:]:
        if maxY:
            if pair[1] >= pareto_front[-1][1]:
                pareto_front.append(pair)
        else:
            if pair[1] <= pareto_front[-1][1]:
                pareto_front.append(pair)
    
    '''Plotting process'''
    plt.scatter(Xs,Ys, alpha=0.4)
    pf_X = [pair[0] for pair in pareto_front]
    pf_Y = [pair[1] for pair in pareto_front]
    plt.plot(pf_X, pf_Y, color='red')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
    
    return pareto_front
```

## Enhanced Parto frontier wrangling

The above code is great for just plotting, but we'd like to extract the associated fields for the Pareto frontier.  So I'll have to redo that Pareto function to work on the entire dataframe.

```python
last_gen_copy = pd.read_csv('last_gen.csv')
```

```python
# Add a new field to filter on for Pareto dominance
last_gen_copy = last_gen_df.copy()
```

```python
last_gen_copy['pareto'] = False
```

```python
last_gen_copy.sort_values(by=['force_fitness','energy_fitness'], inplace=True, ignore_index=True)
```

```python
last_gen_copy.loc[0,'pareto'] = True
```

```python
last_gen_copy[last_gen_copy.pareto].to_csv('just_pareto_front.csv', index=False)
```

```python
# Track the index of the last Pareto frontier neighbor comparison
last_pareto_neighbor = 0
```

```python
for row in range(1, len(last_gen_copy)):
    # print(f'{last_gen_copy.iloc[row]}')
    if last_gen_copy.loc[row, 'energy_fitness'] <= last_gen_copy.loc[last_pareto_neighbor, 'energy_fitness']:
        last_gen_copy.loc[row, 'pareto'] = True
        last_pareto_neighbor = row
    else:
        last_gen_copy.loc[row, 'pareto'] = False
```
